```python
import pandas as pd

```

```python
for period in  ["1221", "0322", "0622", "0922", "1222"]:
    yahoo_old = pd.read_excel(f"data/yahoo/adj_price_yahoo_comp_{period}.xlsx")
    yahoo_new = pd.read_excel(f"data/yahoo/adj_price_yahoo_comp_{period}_new.xlsx")
    yahoo_old.set_index(yahoo_old.Date, drop=True, inplace=True)
    yahoo_new.set_index(yahoo_new.Date, drop=True, inplace=True)
    yahoo_new = yahoo_new.reindex(yahoo_old.index)
    common = yahoo_new.columns.intersection(yahoo_old.columns)

    y1 = yahoo_new[common].copy()
    y2 = yahoo_old[common].copy()
    diff = y1.compare(y2)        # default shows self vs other
    print(period)
    print(diff)
```
