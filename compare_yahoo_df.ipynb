```python
for period in  ["1221", "0322", "0622", "0922", "1222"]:
    yahoo_old = pd.read_excel(f"data/yahoo/adj_price_yahoo_comp_{period}.xlsx")
    yahoo_new = pd.read_excel(f"data/yahoo/adj_price_yahoo_comp_{period}_new.xlsx")
    yahoo_old.set_index(yahoo_old.Date, drop=True, inplace=True)
    yahoo_new.set_index(yahoo_new.Date, drop=True, inplace=True)
    yahoo_new = yahoo_new.reindex(yahoo_old.index)
    common = yahoo_new.columns.intersection(yahoo_old.columns)

    y1 = yahoo_new[common].copy()
    y2 = yahoo_old[common].copy()
    diff = y1.compare(y2)        # default shows self vs other
    print(period)
    print(diff)
```

```python
import pandas as pd

# ---------- 1) PRICE SHEET ----------
price_div_old = pd.read_excel("data/lseg/prices_dividends/price_div_comp_1222.xlsm",
                              sheet_name='CLOSE PRICE', header=4)

price_div_next_3m = pd.read_excel("data/lseg/prices_dividends/price_div_comp_1222_next3m.xlsm",
                                  sheet_name='CLOSE PRICE', header=4)

price_div_next_3m = price_div_next_3m.loc[
    (price_div_next_3m['Code'] > "2023") &
    (price_div_next_3m['Code'] <= "2023-03-31")
]

price_div = pd.concat([price_div_old, price_div_next_3m], axis=0)


# ---------- 2) DIV RATE SHEET ----------
div_rate_old = pd.read_excel("data/lseg/prices_dividends/price_div_comp_1222.xlsm",
                             sheet_name='DIV RATE', header=4)

div_rate_next_3m = pd.read_excel("data/lseg/prices_dividends/price_div_comp_1222_next3m.xlsm",
                                 sheet_name='DIV RATE', header=4)

div_rate_next_3m = div_rate_next_3m.loc[
    (div_rate_next_3m['Code'] > "2023") &
    (div_rate_next_3m['Code'] <= "2023-03-31")
]

div_rate = pd.concat([div_rate_old, div_rate_next_3m], axis=0)


# ---------- 3) DIV DATE SHEET ----------
div_date_old = pd.read_excel("data/lseg/prices_dividends/price_div_comp_1222.xlsm",
                             sheet_name='DIV DATE', header=4)

div_date_next_3m = pd.read_excel("data/lseg/prices_dividends/price_div_comp_1222_next3m.xlsm",
                                 sheet_name='DIV DATE', header=4)

div_date_next_3m = div_date_next_3m.loc[
    (div_date_next_3m['Code'] > "2023") &
    (div_date_next_3m['Code'] <= "2023-03-31")
]

div_date = pd.concat([div_date_old, div_date_next_3m], axis=0)


price_div.to_excel("data/lseg/prices_dividends/price_1222_new.xlsx")
div_rate.to_excel("data/lseg/prices_dividends/div_rate_1222_new.xlsx")
div_date.to_excel("data/lseg/prices_dividends/div_date_1222_new.xlsx")
```
